# Sales Predictive Modeling: Model Development

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# load model dataset
model_data = pd.read_csv("bigmart_model_dataset.csv", index_col=0)
model_data.head()
```

ML Approach: 
+ Model Training
+ Model Building
+ Model Selection
+ Model Evaluation
+ Error Analysis

### Model Training

```python
# define function to train the model
def model_training(): 
  spmX = model_data.drop(columns=["Item_Outlet_Sales"], axis=1)
  spmY = model_data["Item_Outlet_Sales"]

  spm_xtrain, spm_xtest, spm_ytrain, spm_ytest = train_test_split(spmX, spmY, test_size=0.25, random_state=1234)
  training_set = spm_xtrain, spm_ytrain
  testing_set = spm_xtest, spm_ytest
  return [training_set, testing_set]
```

```python
# train the model
spm_training, spm_testing = model_training()
spm_xtrain, spm_ytrain = spm_training
spm_xtest, spm_ytest = spm_testing
```

```python
# build linear regression model
lin_model = LinearRegression()
lin_model.fit(spm_xtrain, spm_ytrain)
```
